#### ระบุชื่อของเส้น => legend   *****ทั้งหมดนี่เขียนใน box legend*****
1 กำหนดตำเเหน่งของ legend ใช้ได้ทั้งคู่ Format String, Code  => loc

- Format String     => Code
- 'best' (default)  =>  0
- 'upper right'     =>  1
- 'upper left'      =>  2
- 'lower left'      =>  3
- 'lower right'     =>  4
- 'right'           =>  5
- 'center left'     =>  6
- 'center right'    =>  7
- 'lower center'    =>  8
- 'upper center'    =>  9
- 'center'          =>  10

2 ขนาดของของชื่อของเส้นใน legend => fontsize
- ค่า xx-small, x-small, small, medium, large, x-large, xx-large
- ค่า ตัวเลข 10, 20 อะไรก็ได้

3 facecolor, egdcolor
- facecolor คือ สีของพื้นหลังของ legend ชื่อเต็มหรือย่อก็ได้
- edgecolor คือ สีของเส้นของ legend ชื่อเต็มหรือย่อก็ได้

4 title, title_fontsize
- title คือ title ของ legend
- title_fontsize คือ ขนาดของ title ของ legend
- ค่า xx-small, x-small, small, medium, large, x-large, xx-large หรือเป็นตัวเลขก็ได้

5 borderpad, labelspacing เอาไว้ปรับขนาดของ legend
* borderpad คือ ขนาด ความใหญ่ของ legend dafault = 0.4
* labelspacing คือ ระยะห่างระหว่างตัวอักษร default = 0.5

### กราฟวงกลม => pie
เหมาะกับเเสดงข้อมูลที่เป็นสัดส่วน หรือเน้นส่วนที่สนใจ
+ labels คือ ใส่ชื่อให้กับสัดส่วนของข้อมูล
+ autopct คือ เเสดงสัดส่วนว่ามีกี่ % => autopct='%.2f%%'

```python
a = [12, 30, 45, 60, 40]
lang=['A', 'B', 'C', 'D', 'E']
plt.pie(x=a, labels=lang, autopct='%.1f%%')
plt.show()
```

```python
a = [12, 30, 45, 60, 40]
lang=['A', 'B', 'C', 'D', 'E']
plt.pie(x=a, labels=lang, autopct='%.2f%%')
plt.show()
```

```python
x = 2002.52855555
print(f'{x:.2f}')
```

#### เเต่งกราฟวงกลม
* x คือ data ที่ต้องใส่ข้อมูลลงไป
* labels คือ ชื่อของกลุ่มนั้นๆ
* autopct คือ ให้เเสดงสัดส่วนของพื้นที่กี่ %
* colors คือ กำหนดสีเอง
* shadow = True คือ ทำให้มีเงาออกมา
* explode คือ เอาส่วนนั้นออกมาจากกราฟวงกลม จำนวนข้อมูล=explode เเล้วกำหนดว่าให้ห่างออกมาเท่าไร เรียงตามตำเเหน่ง
* startangle คือ หมุนกราฟเป็น องศา

```python
a = [12, 30, 45, 60, 40]
lang = ['A', 'B', 'C', 'D', 'E']
colors1 = ['r', 'g', 'b', 'purple', 'yellow']
exp = [0,0.5,0,0,0.2]
plt.pie(x=a,labels=lang, 
colors=colors1, autopct='%.2f%%', shadow=True, explode=exp, startangle=45)
plt.show()
```

### กราฟเเท่ง => bar
* x คือ ชื่อของเเต่ละเเท่ง
* height คือ ขอมูลมีค่าเท่าไร
* color คือ ใส่สีของเเต่ละเเท่ง จำนวนก็ต้องเท่ากันด้วย
* width คือ ความกว้างของ bar  default = 0.8
* edgecolor คือ สีของเส้นขอบ
* linewidth คือ ความหนาเส้นขอบ

กราฟเเท่งเเนวนอน => barh
* y คือ ชื่อเเต่ละเเท่ง
* color คือ สี
* edgecolor คือ สีเส้นขอบ
* width คือ ค่าที่ใส่เข้าไป เหมือน height ของ bar
* linewidth คือ ความหนาของเส้นขอบ 
* height คือ ความหนาของเเท่งเเทน

```python
aa = ['A', 'B', 'C']
bb = [34,56,87]
cc = ['red', 'g', 'gold']
plt.bar(x=aa, height=bb, color=cc, width=0.5 , edgecolor = 'k', linewidth = 5)
plt.xlabel('5678')
plt.show()
```

```python
aa = ['A', 'B', 'C']
bb = [34,56,87]
cc = ['red', 'g', 'gold']
plt.barh(y=aa, color = cc, edgecolor = 'k', width=bb, linewidth = 5, height=[0.5,0.1,0.5])
plt.ylabel('5678')
plt.show()
```

### stackbar|stackplot

```python
room = ['A', 'B', 'C']
boy = [10, 15, 5]
girls = [20, 7, 15]

plt.bar(x=room, height=boy, width=0.5, label = 'Boy', color= 'orange')
plt.bar(x=room, height=girls, width=0.5, label = 'Girls', color = 'pink'
        ,bottom=boy)
        # bottom คือ ให้ boy มาอยู่ข้างล่าง
        # กำหนด label ใน bar เเล้วมาใช้คำสั่ง legend ให้เเสดงผลออกมา
plt.legend()
```

```python
room = ['A', 'B', 'C']
boy = [10, 15, 5]
girls = [20, 7, 15]

plt.stackplot(room, boy, girls, labels=['boy', 'girls'], colors = ['g', 'r'])
# กำหนด label ใน bar เเล้วมาใช้คำสั่ง legend ให้เเสดงผลออกมา
plt.legend()
plt.show()
```
